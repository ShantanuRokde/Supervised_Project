```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'c:\\Users\\SHANTANU\\Downloads\\water_dataX.csv', encoding='unicode_escape')
df
```

```python
def convert_to_numeric_Value(df):
    num_col = df.shape[1]
    for index in range(3, num_col):
        col_name = df.iloc[:, index].name
        print(col_name)
        df[col_name] = pd.to_numeric(df[col_name], errors="coerce")
    return df

df = convert_to_numeric_Value(df)
df.dtypes
```

```python
df.drop(df.columns[[0,1,2,11]],axis=1,inplace=True)
df.isnull().sum()
```

```python
for i in df.columns:
    df[i]=df[i].fillna(value=df[i].mean())
```

```python
df.isnull().sum()
```

```python
start=2
end=1991
do= df.iloc [start:end ,1].astype(np.float64)
value=0
ph = df.iloc[ start:end,2]  
co = df.iloc [start:end ,3].astype(np.float64)
tc=df.iloc [2:end ,7].astype(np.float64)
bod = df.iloc [start:end ,4].astype(np.float64)
na= df.iloc [start:end ,5].astype(np.float64)
na.dtype
```

```python
df=pd.concat([do,ph,co,bod,na,tc],axis=1)
df. columns = ['do','ph','co','bod','na','tc']
df
```

```python
#calulation of Ph
df['pH']=df.ph.apply(lambda x: (100 if (8.5>=x>=7)  
                                 else(80 if  (8.6>=x>=8.5) or (6.9>=x>=6.8) 
                                      else(60 if (8.8>=x>=8.6) or (6.8>=x>=6.7) 
                                          else(40 if (9>=x>=8.8) or (6.7>=x>=6.5)
                                              else 0)))))
```

```python
#calculation of dissolved oxygen
df['do']=df.do.apply(lambda x:(100 if (x>=6)  
                                 else(80 if  (6>=x>=5.1) 
                                      else(60 if (5>=x>=4.1)
                                          else(40 if (4>=x>=3) 
                                              else 0)))))
```

```python
#calculation of total coliform
df['co']=df.tc.apply(lambda x:(100 if (5>=x>=0)  
                                 else(80 if  (50>=x>=5) 
                                      else(60 if (500>=x>=50)
                                          else(40 if (10000>=x>=500) 
                                              else 0)))))
```

```python
#calc of B.D.O
df['bdo']=df.bod.apply(lambda x:(100 if (3>=x>=0)  
                                 else(80 if  (6>=x>=3) 
                                      else(60 if (80>=x>=6)
                                          else(40 if (125>=x>=80) 
                                              else 0)))))
```

```python
#calculation of electrical conductivity
df['ec']=df.co.apply(lambda x:(100 if (75>=x>=0)  
                                 else(80 if  (150>=x>=75) 
                                      else(60 if (225>=x>=150)
                                          else(40 if (300>=x>=225) 
                                              else 0)))))
```

```python
#Calulation of nitrate
df['na']=df.na.apply(lambda x:(100 if (20>=x>=0)  
                                 else(80 if  (50>=x>=20) 
                                      else(60 if (100>=x>=50)
                                          else(40 if (200>=x>=100) 
                                              else 0)))))

df.head()
df.dtypes
df
```

```python
df['pH']=df.pH * 0.165
df['do']=df.do * 0.281
df['bod']=df.bdo * 0.234
df['ec']=df.ec* 0.009
df['na']=df.na * 0.028
df['co']=df.co * 0.281
df['wqi']=df.pH+df.do+df.bod+df.ec+df.na+df.co
df.head()
```
